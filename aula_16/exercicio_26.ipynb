##  Execute o algoritmo k-means para a base de dados abaixo, considerando k=2 para os centroides abaixo descritos. Ao final, informe os centroides resultantes e o SSE de cada grupo.

<div style="display: flex;">
    
    
</div>





```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
pontos = np.array([[1, 1], [1, 2], [2, 1], [3, 2], [3, 4], [4, 4]])

centroides_iniciais = np.array([[3, 1], [1, 4]])

kmeans = KMeans(n_clusters=2, init=centroides_iniciais, n_init=1, max_iter=300, random_state=42)

kmeans.fit(pontos)

# Exibir os centroides finais
print("Centroides finais:")
print(kmeans.cluster_centers_)
```

##  Centroides finais: 
C1 (X = 1.75, Y = 1.5)\
C2 (X=3.50,Y=4.0) SSE=3.74(C1)+0.50(C2)=4,24
