## Tarefas a serem realizadas:
### Unificar os arquivos: Combine os dados de pokedex_1.csv e pokedex_2.json em um único DataFrame, garantindo que as informações estejam corretamente alinhadas.

```python
pokedex = pd.read_csv('pokedex_(Update_04.21).csv')

pokedex.columns
```

### Remoção de duplicatas: Verifique a existência de registros duplicados no DataFrame consolidado e elimine-os para garantir a unicidade dos dados.

```python
remove_duplicates = pokedex.drop_duplicates()
remove_duplicates
```

### Remover colunas irrelevantes: Analise as colunas presentes e exclua aquelas que não são relevantes para a análise

```python
remover_irrelevantes = remove_duplicates.drop(columns = ['Unnamed: 0', 'japanese_name', 'german_name'])
remover_irrelevantes
```

### Tratar valores ausentes: Identifique valores nulos ou ausentes e aplique as ações necessárias para corrigi-los (preenchimento ou remoção).

```python
valores_ausentes = remover_irrelevantes.isnull().sum()
valores_ausentes
```

```python
# Tratamento de valores ausentes em remover_irrelevantes, mantendo todas as colunas
dataframe_tratado = remover_irrelevantes.fillna({'type_2': 'Unknown', 'ability_2': 'Unknown', 
                                                    'ability_hidden': 'Unknown', 'catch_rate': 'Unknown', 
                                                    'base_friendship': 'Unknown', 'base_experience': 'Unknown', 
                                                    'growth_rate': 'Unknown', 'egg_type_1': 'Unknown', 
                                                    'egg_type_2': 'Unknown', 'percentage_male': 'Unknown', 
                                                    'egg_cycles': 'Unknown'})

dataframe_tratado
```

### Filtrar Pokémon proibidos: Exclua do conjunto de dados todos os Pokémon que pertencem às seguintes categorias:
• Lendários e Sub-lendários\
• Míticos\
• Pokémon com Mega Evoluções


```python
# Filtrar Pokémon proibidos

# Agora, aplique o filtro de Pokémon proibidos
df_proibidos = remover_irrelevantes[~remover_irrelevantes['status'].isin(['Legendary', 'Sub Legendary', 'Mythical']) & 
                                    ~remover_irrelevantes['name'].str.contains('Mega ', na=False)]
df_proibidos

```
