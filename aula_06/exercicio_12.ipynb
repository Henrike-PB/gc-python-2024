 # Exercício 12 - Inner Join Objetivo:
## Unir os DataFrames de livros e empréstimos usando o método inner join e mostrar apenas os livros que foram emprestados.

# 1. Carregue os arquivos livros.csv e emprestimos.csv em DataFrames separados.

```python
import pandas as pd

df_livros = pd.read_csv("livros.csv")
df_livros

df_emprestimos = pd.read_csv('emprestimos.csv')
df_emprestimos    
```

# 2. Realize um inner join entre as tabelas, unindo-as pela coluna id_livro.

```python
df_emprestimos_livros = pd.merge(df_livros, df_emprestimos, on='id_livro', how='inner')
df_emprestimos_livros
```

# 3. Exiba o resultado, mostrando as colunas: titulo, autor, nome_leitor, data_emprestimo e data_devolucao.

```python
df_emprestimos_livros[['titulo', 'autor', 'nome_leitor', 'data_emprestimo', 'data_devolucao']]
df_emprestimos_livros
```
