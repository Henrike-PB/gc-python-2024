```python
dataset_penguins.describe()
```

```python
# Selecionar apenas as colunas numéricas para calcular a média
numeric_columns = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
mean_values = dataset_penguins.groupby('species')[numeric_columns].mean()
mean_values
```

```python
import plotly.express as px

# Gráfico de boxplot para comparar o peso corporal entre as espécies
fig_weight = px.box(dataset_penguins, x='species', y='body_mass_g', title='Distribuição do Peso Corporal por Espécie')
fig_weight.show()
```

```python
# Gráfico de boxplot para comparar o comprimento do bico entre as espécies
fig_bill_length = px.box(dataset_penguins, x='species', y='bill_length_mm', title='Distribuição do Comprimento do Bico por Espécie')
fig_bill_length.show()
```

```python
# Gráfico de boxplot para comparar o comprimento da asa entre as espécies
fig_flipper_length = px.box(dataset_penguins, x='species', y='flipper_length_mm', title='Distribuição do Comprimento da Asa por Espécie')
fig_flipper_length.show()
```

### • Crie gráficos para visualizar as diferenças entre as espécies de pinguins com base nas características físicas. Use gráficos de dispersão, boxplot e histogramas para mostrar as distribuições e comparações entre as variáveis.

```python
# Gráfico de dispersão para comprimento do bico vs. profundidade do bico
fig_scatter_bill = px.scatter(dataset_penguins, x='bill_length_mm', y='bill_depth_mm', color='species',
                              title='Comprimento do Bico vs. Profundidade do Bico')
fig_scatter_bill.show()
```

```python
# Gráfico de dispersão para comprimento da asa vs. peso corporal
fig_scatter_flipper = px.scatter(dataset_penguins, x='flipper_length_mm', y='body_mass_g', color='species',
                                 title='Comprimento da Asa vs. Peso Corporal')
fig_scatter_flipper.show()
```

```python
# Gráfico de boxplot para comparar o peso corporal entre as espécies
fig_box_weight = px.box(dataset_penguins, x='species', y='body_mass_g', title='Distribuição do Peso Corporal por Espécie')
fig_box_weight.show()
```

```python
# Gráfico de boxplot para comparar o comprimento do bico entre as espécies
fig_box_bill_length = px.box(dataset_penguins, x='species', y='bill_length_mm', title='Distribuição do Comprimento do Bico por Espécie')
fig_box_bill_length.show()
```

```python
# Gráfico de boxplot para comparar o comprimento da asa entre as espécies
fig_box_flipper_length = px.box(dataset_penguins, x='species', y='flipper_length_mm', title='Distribuição do Comprimento da Asa por Espécie')
fig_box_flipper_length.show()
```

```python
# Histograma para o comprimento do bico
fig_hist_bill_length = px.histogram(dataset_penguins, x='bill_length_mm', color='species', barmode='overlay',
                                    title='Distribuição do Comprimento do Bico por Espécie')
fig_hist_bill_length.show()
```

```python
# Histograma para o comprimento da asa
fig_hist_flipper_length = px.histogram(dataset_penguins, x='flipper_length_mm', color='species', barmode='overlay',
                                       title='Distribuição do Comprimento da Asa por Espécie')

fig_hist_flipper_length.show()
```

## Impacto das Mudanças Climáticas
As mudanças climáticas estão afetando os habitats dos pinguins de várias maneiras:

## Disponibilidade de Alimentos:
O aquecimento dos oceanos e a redução do gelo marinho podem impactar a disponibilidade de krill, um alimento essencial para os pinguins. Isso pode explicar as variações no peso corporal observadas entre as espécies, com algumas espécies sendo mais afetadas pela escassez de alimentos.
## Mudanças no Habitat:
A perda de gelo marinho e as mudanças na cobertura de neve podem alterar os locais de nidificação e reprodução dos pinguins. Espécies que dependem de gelo marinho, como os pinguins Adélie, podem enfrentar desafios maiores em comparação com espécies mais adaptáveis, como os Gentoo.
## Saúde e Sobrevivência:
As mudanças nas condições ambientais podem afetar a saúde e a sobrevivência dos pinguins. A análise dos dados de saúde e dieta pode revelar correlações entre a saúde dos pinguins e as mudanças no ambiente, destacando a importância de monitorar essas populações para entender os impactos a longo prazo.\
